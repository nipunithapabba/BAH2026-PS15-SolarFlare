```python
from sunpy.net import Fido, attrs as a

result = Fido.search(
    a.Time("2024-01-01", "2025-12-31"),
    a.hek.EventType("FL"),
    a.hek.FL.GOESCls >= "B1.0",
    a.hek.FL.GOESCls < "C1.0"
)

flares = result["hek"]
print(f"Found {len(flares)} B-class flares")
print(flares["event_starttime", "event_peaktime", "event_endtime", "fl_goescls"][:15])
```
